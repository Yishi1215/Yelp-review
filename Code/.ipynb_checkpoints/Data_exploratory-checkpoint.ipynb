```python
useful_reviews = len(df[df["useful"]>0])
negative_reviws = len(df[df["stars"]<2])
positive_reviews =len(df[df["stars"]>3])
total_reviews = len(df)

print("Total reviews: {}".format(total_reviews))
print("Useful reviews: {}".format(useful_reviews))
print("Total negative reviews: {}".format(negative_reviws))
print("Total positive reviews: {}".format(positive_reviews))
```

```python
df['year'] = pd.to_datetime(df['date']).map(lambda x: x.strftime("%Y"))

year_useful_sum = df.groupby(['year'])['useful'].sum().reset_index()
year_useful_mean = df.groupby(['year'])['useful'].mean().reset_index()

p1 = figure(title="Year vs. Total Useful Vote",
x_axis_label='Year',
y_axis_label='Total Useful Vote')
p1.line(year_useful_sum.year, year_useful_sum.useful, color="firebrick")

p2 = figure(title="Year vs. Average Useful Vote",
x_axis_label='Year',
y_axis_label='Average Useful Vote')
p2.line(year_useful_mean.year, year_useful_mean.useful, color="firebrick")

output_file("../plot/date_vs_useful.html")
```

```python
df_positive = df[df.pol>0]
k= (' '.join(df_positive['text']))
stopwords = set(STOPWORDS)
wordcloud = WordCloud(width = 1000, height = 500, stopwords=stopwords).generate(k)
plt.figure(figsize=(15,5))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('../plot/positive.jpg')
```

```python
df_negative = df[df.pol<0]
k= (' '.join(df_negative['text']))
stopwords = set(STOPWORDS)
wordcloud = WordCloud(width = 1000, height = 500, stopwords=stopwords).generate(k)
plt.figure(figsize=(15,5))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('../plot/negative.jpg')
```

```python
# conbine the reviews to a single string
text = '\n'.join(df['text'])

# make a wordcloud
d = path.dirname('__file__')

# read the mask image
alice_coloring = np.array(Image.open(path.join(d, "../Other/alice_color.png")))

stopwords = set(STOPWORDS)

wc = WordCloud(background_color="white", max_words=2000, mask=alice_coloring,
               stopwords=stopwords, max_font_size=40, random_state=42)

# generate word cloud
wc.generate(text)

# store to file
wc.to_file(path.join(d, "../Other/alice_wordcloud.png"))

# show
fig1 = plt.gcf()
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure(figsize=(5,15))
# plt.savefig('../plot/review_alice.jpg', dpi=1600)
fig1.savefig('../plot/review_alice.jpg')
```
