```python
df_review_top['textCount'] = df_review_top['text'].str.split().str.len()
df_review_top = df_review_top.sort_values('textCount', ascending = False)
```

```python
df_review_top['year'] = pd.to_datetime(df_review_top['date']).map(lambda x: x.strftime("%Y"))

year_useful_sum = df_review_top.groupby(['year'])['useful'].sum().reset_index()
year_useful_mean = df_review_top.groupby(['year'])['useful'].mean().reset_index()

p1 = figure(title="Year vs. Total Useful Vote",
x_axis_label='Year',
y_axis_label='Total Useful Vote')
p1.line(year_useful_sum.year, year_useful_sum.useful, color="firebrick")

p2 = figure(title="Year vs. Average Useful Vote",
x_axis_label='Year',
y_axis_label='Average Useful Vote')
p2.line(year_useful_mean.year, year_useful_mean.useful, color="firebrick")

show(row(p1,p2))
```
